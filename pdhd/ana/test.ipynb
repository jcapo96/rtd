```python
import pandas as pd
import time, datetime, tqdm
import matplotlib.pyplot as plt
import numpy as np
import pickle, utils
```

```python
tini, tend = datetime.datetime(2024, 7, 15, 0, 0, 0), datetime.datetime(2024, 8, 15, 0, 0, 0)
data, data_err = utils.load_data(tini, tend)
```

```python
fig, axes = plt.subplots(1,1)
axes.set_xlim(datetime.datetime(2024, 7, 28), datetime.datetime(2024, 8, 2))
axes2 = axes.twinx()

conditions = (data["TE0241"]<90)&(data["TE0241"]>87)

axes.plot(data.loc[conditions].index.to_numpy(), data.loc[conditions]["TE0241"].to_numpy(), label="TE0241")
axes.set_ylabel("Temperature (K)")
axes.legend(loc="lower left")
axes2.plot(data.index.to_numpy(), data["Heinz_VRaw"].to_numpy(), ".", color="tab:red")
axes2.set_ylabel("Heinzinger HV (V)")


# data["TE0001"].loc[(data["TE0001"]<90)&(data["TE0001"]>87)].plot()
# data["Heinz_VRaw"].plot()
```
